```python
import pandas as pd
from sklearn.impute import SimpleImputer
import numpy as np

df = pd.DataFrame(pd.read_excel("data.xlsx"))

#
l = list(df.columns[52:])
df = df.drop(l, axis = 1)

#clean data
df.iloc[:, 41:] = df.iloc[:, 41:].replace(r'^\s*$', np.nan, regex=True)

#try different strategy
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer = imputer.fit(df.iloc[:, 41:])
df.iloc[:, 41:] = imputer.fit_transform(df.iloc[:, 41:])
```

```python
#analyze data
# train-test split
#train with logistic regression

#part 1 df
df_log = df.iloc[:, 41:]

#change death to binary variable
temp = df.iloc[:,15]
temp = temp.fillna(0)
for i in range(len(temp)):
    if temp[i] > 0:
        temp[i] = 1
        
df_log["DEATH [d from CT]"] = temp
```

```python
from sklearn.model_selection import train_test_split

#train test split
train, test = train_test_split(df_log, test_size=0.2)
train_x = train.iloc[:,1:-1]
#train_x = train[["VAT Area (cm2)", "VAT/SAT     Ratio", "AoCa        Agatston"]]
train_y = train.iloc[:,-1]
test_x = test.iloc[:,1:-1]
#test_x = test[["VAT Area (cm2)", "VAT/SAT     Ratio", "AoCa        Agatston"]]
test_y = test.iloc[:,-1] 
train_x
```

```python
df_log.columns
```

```python
#log model
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import r2_score
from sklearn import metrics
from sklearn.preprocessing import PolynomialFeatures

pipe = Pipeline([('sc', StandardScaler()),
                    ('pca', PCA(n_components=3)),
                    ('t', PolynomialFeatures(degree=2)),
                    ('clf', LogisticRegression(multi_class='multinomial'))
                    ])
pipe.fit(train_x, train_y)

#high because of less input
pipe.score(test_x, test_y)

#negative bad
#r2_score(list(test_y), pipe.predict(test_x))
metrics.f1_score(list(test_y), list(pipe.predict(test_x)))
#sum(list(pipe.predict(test_x)))
```
